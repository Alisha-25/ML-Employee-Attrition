```python
df.drop(['EmployeeCount','Over18','StandardHours'], axis=1, inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
# now we will assume all columns in our data which are object columns are nominal
# so we will do one-hot encoding bcoz of nominal

df_ohe = pd.get_dummies(df)
```

```python
df_ohe.shape
```

```python
df_ohe.fillna(df_ohe.median(), inplace=True)
```

```python
df_ohe.columns
```

```python
df_ohe.nunique()
```

# X & Y split

```python
X = df_ohe.drop(['Attrition_No','Attrition_Yes'],axis=1)
X
```

```python
Y = df_ohe['Attrition_Yes']
Y
```

```python
X.shape, Y.shape
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=7)
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

# standardize the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
sd = StandardScaler()                       # StandardScaler is algo here
sd.fit(X_train, Y_train)                    # fit() trains an algo
# learns mean and std_deviation from training data
```

```python
# after standardization, mean is 0 and std_dev is 1, always .... and data is centered around mean
```

```python
X_train.describe()
# can see standardscaler is nota pplied yet, we have just fitted model, so now use transform()
```

```python
# X_train_std = sd.transform(X_train)
# this changes the data based on z-score i.e. (X-mean)/std_dev
```

```python
# X_train_std.describe()
# gives AttributeError: 'numpy.ndarray' object has no attribute 'describe'
```

```python
X_train.loc[:,:] = sd.transform(X_train)
# this changes the data based on z-score i.e. (X-mean)/std_dev
```

```python
X_train.describe()
```

```python
X_test.describe()
```

```python
X_test.loc[:,:] = sd.transform(X_test)
```
